```python
# =========================
# STEP 1 - IMPORT
# =========================
import pandas as pd
from sklearn.preprocessing import LabelEncoder
!pip install xgboost
import numpy as np

# =========================
# STEP 2 - LOAD DATA
# =========================
train = pd.read_csv('Train.csv')
test = pd.read_csv('Test.csv')

test_ids = test['Place_ID X Date']

# =========================
# STEP 3 - CLEAN TRAIN
# =========================
drop_cols = ['target_min', 'target_max', 'target_variance', 'target_count']
train = train.drop(columns=drop_cols)

y = train['target']
X = train.drop(columns=['target'])

# =========================
# STEP 4 - DROP ID COLUMN
# =========================
X = X.drop(columns=['Place_ID X Date'])
test = test.drop(columns=['Place_ID X Date'])

# =========================
# STEP 5 - ENCODE Place_ID
# =========================
X['Place_ID'] = X['Place_ID'].astype(str)
test['Place_ID'] = test['Place_ID'].astype(str)

le = LabelEncoder()
all_ids = pd.concat([X['Place_ID'], test['Place_ID']])
le.fit(all_ids)

X['Place_ID'] = le.transform(X['Place_ID'])
test['Place_ID'] = le.transform(test['Place_ID'])

# =========================
# STEP 6 - DATE FEATURES (STRONGER)
# =========================
X['Date'] = pd.to_datetime(X['Date'])
test['Date'] = pd.to_datetime(test['Date'])

for df in [X, test]:
    df['day'] = df['Date'].dt.day
    df['month'] = df['Date'].dt.month
    df['dayofweek'] = df['Date'].dt.dayofweek
    df['weekofyear'] = df['Date'].dt.isocalendar().week.astype(int)

X = X.drop(columns=['Date'])
test = test.drop(columns=['Date'])

# =========================
# STEP 7 - CREATE WIND SPEED (VERY IMPORTANT)
# =========================
for df in [X, test]:
    df['wind_speed'] = np.sqrt(
        df['u_component_of_wind_10m_above_ground']**2 +
        df['v_component_of_wind_10m_above_ground']**2
    )

# =========================
# STEP 8 - FILL MISSING
# =========================
X = X.fillna(X.mean())
test = test.fillna(test.mean())

# =========================
# STEP 9 - MODEL (STRONGER SETTINGS)
# =========================
from xgboost import XGBRegressor

model = XGBRegressor(
    n_estimators=700,
    learning_rate=0.05,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

model.fit(X, y)
preds = model.predict(test)

model.fit(X, y)

# =========================
# STEP 10 - PREDICT
# =========================
preds = model.predict(test)

# =========================
# STEP 11 - SUBMIT
# =========================
submission = pd.DataFrame({
    'Place_ID X Date': test_ids,
    'target': preds
})

submission.to_csv('submission_XGBoost.csv', index=False)

print("DONE - SUBMIT THIS")
```
