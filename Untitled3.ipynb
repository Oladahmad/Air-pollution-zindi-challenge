```python
# =========================
# STEP 1 - IMPORT
# =========================
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# install models
!pip install xgboost lightgbm

from xgboost import XGBRegressor
from lightgbm import LGBMRegressor

# =========================
# STEP 2 - LOAD DATA
# =========================
train = pd.read_csv('Train.csv')
test = pd.read_csv('Test.csv')

test_ids = test['Place_ID X Date']

# =========================
# STEP 3 - CLEAN TRAIN
# =========================
drop_cols = ['target_min', 'target_max', 'target_variance', 'target_count']
train = train.drop(columns=drop_cols)

y = train['target']
X = train.drop(columns=['target'])

# =========================
# STEP 4 - DROP ID
# =========================
X = X.drop(columns=['Place_ID X Date'])
test = test.drop(columns=['Place_ID X Date'])

# =========================
# STEP 5 - ENCODE Place_ID
# =========================
X['Place_ID'] = X['Place_ID'].astype(str)
test['Place_ID'] = test['Place_ID'].astype(str)

le = LabelEncoder()
all_ids = pd.concat([X['Place_ID'], test['Place_ID']])
le.fit(all_ids)

X['Place_ID'] = le.transform(X['Place_ID'])
test['Place_ID'] = le.transform(test['Place_ID'])

# =========================
# STEP 6 - DATE FEATURES
# =========================
X['Date'] = pd.to_datetime(X['Date'])
test['Date'] = pd.to_datetime(test['Date'])

for df in [X, test]:
    df['day'] = df['Date'].dt.day
    df['month'] = df['Date'].dt.month
    df['dayofweek'] = df['Date'].dt.dayofweek
    df['weekofyear'] = df['Date'].dt.isocalendar().week.astype(int)

X = X.drop(columns=['Date'])
test = test.drop(columns=['Date'])

# =========================
# STEP 7 - WIND FEATURE
# =========================
for df in [X, test]:
    df['wind_speed'] = np.sqrt(
        df['u_component_of_wind_10m_above_ground']**2 +
        df['v_component_of_wind_10m_above_ground']**2
    )

# =========================
# STEP 8 - FILL MISSING
# =========================
X = X.fillna(X.mean())
test = test.fillna(test.mean())

# =========================
# STEP 9 - XGBOOST (BEST)
# =========================
xgb_model = XGBRegressor(
    n_estimators=1200,
    learning_rate=0.03,
    max_depth=7,
    subsample=0.85,
    colsample_bytree=0.85,
    reg_alpha=0.1,
    reg_lambda=1.0,
    random_state=42
)

xgb_model.fit(X, y)
xgb_preds = xgb_model.predict(test)

# =========================
# STEP 10 - LIGHTGBM
# =========================
lgb_model = LGBMRegressor(
    n_estimators=900,
    learning_rate=0.04,
    max_depth=9,
    num_leaves=60,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

lgb_model.fit(X, y)
lgb_preds = lgb_model.predict(test)

# =========================
# STEP 11 - ENSEMBLE (IMPORTANT)
# =========================
# Weighted average (XGBoost slightly stronger)
final_preds = (0.6 * xgb_preds) + (0.4 * lgb_preds)

# =========================
# STEP 12 - SUBMISSION
# =========================
submission = pd.DataFrame({
    'Place_ID X Date': test_ids,
    'target': final_preds
})

submission.to_csv('submission_ensemble.csv', index=False)

print("DONE - SUBMIT THIS")
```
